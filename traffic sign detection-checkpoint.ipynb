```python
import tensorflow as tf

dataset = r"C:\Users\gowri\OneDrive\Desktop\traffic sign detection\traffic_Data\DATA"  # Set correct dataset path

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    dataset,
    validation_split=0.2,
    subset='training',
    seed=123,
    image_size=(224, 224),
    batch_size=32
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    dataset,
    validation_split=0.2,
    subset='validation',
    seed=123,
    image_size=(224, 224),
    batch_size=32
)
```

```python
class_numbers = train_ds.class_names
class_names = []
for i in class_numbers:
    class_names.append(labelfile['Name'][int(i)])
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(25):
        ax = plt.subplot(5, 5, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")

plt.show()
```

```python
import tensorflow as tf

data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal", input_shape=(224, 224, 3)),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomFlip("horizontal_and_vertical")  # No 'mode' argument needed
])
```

```python
model = Sequential()
model.add(data_augmentation)
model.add(Rescaling(1./255))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(256, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(Dense(len(labelfile), activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Set callback functions to early stop training
mycallbacks = [EarlyStopping(monitor='val_loss', patience=5)]
history = model.fit(train_ds,
                 validation_data=val_ds,
                 epochs=20,
                 callbacks=mycallbacks)
```

```python
# Print all available metrics from history
print("Available metrics:", history.history.keys())

# Get final training metrics
train_loss = history.history['loss'][-1]
print("Final Training Loss:", train_loss)

# Get final validation loss (if available)
if 'val_loss' in history.history:
    val_loss = history.history['val_loss'][-1]
    print("Final Validation Loss:", val_loss)

# If accuracy is tracked, show both
if 'accuracy' in history.history:
    train_acc = history.history['accuracy'][-1]
    print("Final Training Accuracy:", train_acc)

if 'val_accuracy' in history.history:
    val_acc = history.history['val_accuracy'][-1]
    print("Final Validation Accuracy:", val_acc)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss')
if 'val_loss' in history.history:
    plt.plot(history.history['val_loss'], label='Validation Loss')

if 'accuracy' in history.history:
    plt.plot(history.history['accuracy'], label='Training Accuracy')

if 'val_accuracy' in history.history:
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')

plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Metric Value")
plt.title("Training & Validation Metrics")
plt.show()
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False  # Freeze base
```

```python
inputs = keras.Input(shape=(224, 224, 3))
x = data_augmentation(inputs)
x = Rescaling(1./255)(x)

x = base_model(x, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)

x = Dense(256, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.2)(x)

outputs = Dense(len(labelfile), activation='softmax')(x)

model2 = keras.Model(inputs, outputs)
model2.summary()
```

```python
model2.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history2 = model2.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20,
    callbacks=[EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True)]
)
```

```python
import matplotlib.pyplot as plt

def plot_history(history, title_prefix=""):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs_range = range(1, len(acc) + 1)

    plt.figure(figsize=(14, 5))

    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Train Accuracy')
    plt.plot(epochs_range, val_acc, label='Val Accuracy')
    plt.title(f'{title_prefix} Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Train Loss')
    plt.plot(epochs_range, val_loss, label='Val Loss')
    plt.title(f'{title_prefix} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()

# BEFORE finetuning
plot_history(history2, "Before Fine-Tuning")
```

```python
base_model.trainable = True  # Unfreeze

# Train only last ~30 layers
for layer in base_model.layers[:-30]:
    layer.trainable = False

model2.compile(
    optimizer=Adam(learning_rate=1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_ft = model2.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=[EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)]
)
```

```python
# AFTER finetuning
plot_history(history_ft, "After Fine-Tuning")
```

```python
import numpy as np
import pandas as pd

def get_best_metrics(history):
    return {
        "Best Train Accuracy": np.max(history.history['accuracy']),
        "Best Val Accuracy": np.max(history.history['val_accuracy']),
        "Lowest Train Loss": np.min(history.history['loss']),
        "Lowest Val Loss": np.min(history.history['val_loss']),
    }

before_metrics = get_best_metrics(history2)
after_metrics = get_best_metrics(history_ft)

comparison_df = pd.DataFrame([before_metrics, after_metrics],
                             index=["Before Fine-Tuning", "After Fine-Tuning"])

print(comparison_df)
```

```python
dataset = r"C:\Users\gowri\OneDrive\Desktop\traffic sign detection\traffic_Data\DATA"  # Set correct dataset path
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    dataset,
    validation_split=0.2,
    subset='training',
    seed=123,
    image_size=(224, 224),
    batch_size=32,
    label_mode="categorical"    # <-- FIX
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    dataset,
    validation_split=0.2,
    subset='validation',
    seed=123,
    image_size=(224, 224),
    batch_size=32,
    label_mode="categorical"    # <-- FIX
)
```

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

base_model = EfficientNetB0(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)

base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.3)(x)

output_layer = Dense(58, activation='softmax')(x)   # <-- FIXED

model = Model(inputs=base_model.input, outputs=output_layer)

model.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20
)
```

```python
import matplotlib.pyplot as plt

# Extract values
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(acc))

plt.figure(figsize=(14, 5))

# -------- Accuracy Plot --------
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.title('Training vs Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# -------- Loss Plot --------
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# ---------- UNFREEZE LAST PART OF BASE MODEL ----------
base_model.trainable = True

fine_tune_at = 200  # unfreeze from layer 200 to end

for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

# ---------- RECOMPILE ----------
model.compile(
    optimizer=Adam(learning_rate=1e-5),   # smaller LR
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ---------- FINE-TUNE ----------
fine_tune_history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10
)
```

```python
import matplotlib.pyplot as plt

# Extract values from fine-tune history
ft_acc = fine_tune_history.history['accuracy']
ft_val_acc = fine_tune_history.history['val_accuracy']

ft_loss = fine_tune_history.history['loss']
ft_val_loss = fine_tune_history.history['val_loss']

epochs_range = range(len(ft_acc))

plt.figure(figsize=(14, 5))

# -------- Fine-Tune Accuracy Plot --------
plt.subplot(1, 2, 1)
plt.plot(epochs_range, ft_acc, label='Fine-Tune Training Accuracy')
plt.plot(epochs_range, ft_val_acc, label='Fine-Tune Validation Accuracy')
plt.title('Fine-Tuning Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# -------- Fine-Tune Loss Plot --------
plt.subplot(1, 2, 2)
plt.plot(epochs_range, ft_loss, label='Fine-Tune Training Loss')
plt.plot(epochs_range, ft_val_loss, label='Fine-Tune Validation Loss')
plt.title('Fine-Tuning Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
